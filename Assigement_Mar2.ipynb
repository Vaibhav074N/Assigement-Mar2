Q1.What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a Python library that is widely used for data visualization. It provides a range of tools for creating static, animated, and interactive visualizations in Python. Some of the common use cases for Matplotlib include data exploration, analysis, and presentation, as well as creating visualizations for scientific publications, reports, and presentations.

Here are five plots that can be created using the Pyplot module of Matplotlib:

1.Line plot - Used to visualize data points connected by straight line segments. Useful for visualizing trends and patterns over time.


2.Scatter plot - Used to visualize the relationship between two variables. Each data point is plotted as a point on a 2D plane, with one variable on the x-axis and the other on the y-axis.


3.Bar plot - Used to compare values between different groups. The height or length of each bar represents the value of the variable being plotted, and the bars are grouped by category.


4.Histogram - Used to visualize the distribution of a single variable. The x-axis represents the range of values for the variable, and the y-axis represents the frequency or density of those values.


5.Pie chart - Used to visualize the proportion of different categories in a dataset. The area of each slice of the pie represents the proportion of the dataset that belongs to that category.

A scatter plot is a type of plot used to display the relationship between two continuous variables. It is a 2-dimensional plot that uses Cartesian coordinates to represent the values of two variables. Each point on the plot represents a single observation, with the x-coordinate representing the value of one variable and the y-coordinate representing the value of the other variable.

```python
import numpy as np
np.random.seed(3)
```

```python
x=3+np.random.normal(0,2,50)
y=3+np.random.normal(0,2,len(x))
```

```python
pt.figure()
pt.xlabel("X-axis")
pt.ylabel("Y-axis")
pt.title("This is X v/s Y Graph")
pt.scatter(x,y)
```

The subplot() function in Matplotlib is used to create multiple plots in a single figure. It allows you to display multiple plots side by side, stacked vertically, or arranged in a grid layout. This can be useful when you want to compare different plots or visualize different aspects of the same data.

The subplot() function takes three arguments: the number of rows, the number of columns, and the index of the plot you want to create

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

fig.suptitle('Four Line Plots')

plt.show()
```

A box plot, also known as a box-and-whisker plot, is a type of graph used to display the distribution of a dataset. It summarizes the data using five statistics: the minimum value, the maximum value, the median, the first quartile (Q1), and the third quartile (Q3). These statistics are represented as a box with whiskers extending from either end.

Box plots are particularly useful when the dataset is large and complex, as they provide a clear and concise summary of the distribution. They can also be used to identify potential issues with the data, such as outliers or skewness. Additionally, box plots can be enhanced with labels, colors, and other features to make them more informative and visually appealing.

```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

fig, ax = plt.subplots()

ax.boxplot(data)

ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_title('Box Plot')

plt.show()
```
